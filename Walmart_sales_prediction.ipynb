# From the exploratory data analysis and visualizations, several key patterns can be observed in the Walmart sales dataset:

Seasonality in Sales
Sales tend to peak during November and December, which is likely driven by holiday seasons such as Thanksgiving and Christmas.
A noticeable decline in sales is observed around January, which may be due to post-holiday effects and discounted or cleared inventory leading to reduced revenue.
Store-wise, Year-wise, and Monthly Trends
Significant variation exists across different stores, years, and months, indicating strong heterogeneity in sales behavior.
This suggests that store-specific and time-based factors play an important role in determining sales performance.
Holiday Impact on Sales
The analysis of holiday vs non-holiday sales indicates that holidays do not consistently lead to significantly higher sales.
Non-holiday periods also show substantial outliers, likely due to promotions, discounts, or special events.
Hence, holidays act more as an opportunity driver rather than a guaranteed sales booster.
Sales are influenced by multiple factors such as:
Type of holiday
Store location
Promotional activities
Seasonal demand patterns
Feature Relationships and Model Implications
The correlation analysis shows generally weak linear relationships between most features and weekly sales.
This is further supported by the low R² value obtained from Multiple Linear Regression, indicating poor explanatory power.
Overall, the dataset exhibits strong non-linear and complex interactions among variables.
